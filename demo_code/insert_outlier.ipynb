```python
import pandas as pd

# Membaca data dari file CSV
detail_out = pd.read_csv("detail_out.csv")

# Menyesuaikan kolom sesuai kebutuhan
detail_out['id_harga_beras'] = detail_out['id']
detail_out['id'] = detail_out['row_id']
detail_out['id_group'] = detail_out['cluster_id']
detail_out = detail_out[['id', 'id_harga_beras', 'id_group']]

# Menghapus duplikat berdasarkan kolom 'id_harga_beras' dan 'id_group', hanya menyisakan satu baris
detail_out_unique = detail_out.drop_duplicates(subset=['id_harga_beras', 'id_group'], keep='first')

# Menampilkan hasil
print(f"Jumlah baris setelah duplikat dihapus: {detail_out_unique.shape[0]}")
print(detail_out_unique)
```

```python
group.to_csv("insert_outlier_group.csv", index=False)
detail_out_unique.to_csv("insert_outlier_detail.csv", index=False)
```

```python
import pandas as pd

# Misalkan df1 adalah DataFrame pertama dan df2 adalah DataFrame kedua
# df1 memiliki kolom 'id_group' dan df2 memiliki kolom 'id'

# Ambil semua nilai 'id_group' dari df1
df1_id_groups = set(detail_out_unique['id_harga_beras'])

# Ambil semua nilai 'id' dari df2 (kolom yang perlu diperiksa keberadaannya)

df = pd.read_csv("harga_beras_rows.csv")
df2_ids = set(df['id'])

# Periksa apakah semua id_group dari df1 ada di df2
missing_ids = df1_id_groups - df2_ids

# Tampilkan hasil
if not missing_ids:
    print("Semua id_group tercatat di df2.")
else:
    print(f"Beberapa id_group tidak tercatat di df2: {missing_ids}")
```
